# ***Método .loc***

# - **Pegar 1 linha**
# - **Pegar 1 linha de acordo com alguma condição**
# - **Pegar linhas e colunas especificadas**
# - **Pegar 1 valor específico**

```python
# Pegar 1 linha específica
'''display(vendas_excel_df.loc[1: 5])''' # Pegando da linha 1 à linha 5


# Pegar linhas que correspondem à uma condição
norte_shopping = vendas_excel_df.loc[vendas_excel_df['ID Loja'] == 'Norte Shopping'] 
# Dentro do .loc[Tenho que informar o nome da coluna onde cada linha seja igual ao nome que eu procuro]

# Criei uma Series da Coluna "Valor Final" para poder levantar um gráfico 
s = pd.Series(norte_shopping['Valor Final'])
s.plot() # Criei um gráfico com os 3924 produtos que foram vendidos

```

 # ***Tratamento de valores vazios*** 

# **São os NaN**
# - Detelar linhas/colunas vazias
# - Deletar linhas que possuem valores vazios
# - Preencher valores vazios com (Média ou Ultimo valor)

```python
# Deletar linhas e colunas COMPLETAMENTE vazias
'''vendas_excel_df = vendas_excel_df.dropna(how='all', axis=1)'''
# dropna com o parâmetro how indica quais linhas e/ou colunas você quer excluir
# Como foi passado o parâmetro "axis=1" eu me referi a todas as COLUNAS com os valores completamete vazios


# Deletar linhas que possuem pelo menos UM valor vazio
'''vendas_excel_df = vendas_excel_df.dropna() # Método PADRÃO para excluir linhas que tenham pelo menos 1 valor vazio'''

# Preencher valores vazios:
# Preencher com a MÉDIA da coluna
vendas_excel_df['Comissão'] = vendas_excel_df['Comissão'].fillna(vendas_excel_df['Comissão'].mean())
display(vendas_excel_df)

# Preencher com o ULTIMO valor
vendas_excel_df = vendas_excel_df.ffill()
# ffill significa que vai preencher valores vazios com o FIRST element que vai preencher o valor da linha com o mesmo valor acima dela
```

# **Groupby**
# **Value Counts**

```python
#Value Counts vai contar quantas vezes determinado item aparece na tabela
transações = vendas_excel_df['ID Loja'].value_counts()
transações.plot(kind='bar')
# Nesse exemplo foi passado como parâmetro o ID LOJA com o nome de cada shopping
# Em seguida foi feito um gráfico dizendo quantas transações foram feitas em cada shopping
```

```python
# Groupby vai agrupar todos os itens de uma determinada coluna
faturamento = vendas_excel_df.groupby('Produto').sum()
faturamento
# Nesse exemplo eu agrupei os items da coluna PRODUTO e somei os valores numéricos restantes de cada coluna
# Por exemplo, agora sabemos que a BERMUDA foi vendida 2061 vezes e foi lucrado R$309.150,00 


# Localizei o item BERMUDA para somar o valor unitário com a quantidade vendida e ver se os valores batem
vendas_excel_df.iloc[391] 
#Os valores bateram
```

```python
# Quero saber SOMENTE a quantidade e o valor final de cada produto
faturamento2 = vendas_excel_df[['Produto', 'Quantidade', 'Valor Final']].groupby('Produto').sum()
faturamento2
# Agora sabemos exatamente quantos produtos foram vendidos e qual foi o faturamento de cada um deles somando em todos os shoppings
```
