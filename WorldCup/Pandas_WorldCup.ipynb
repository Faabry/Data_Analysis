# **Análise de dados de todas as copas do mundo**

# *Importando a base de dados*

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('WorldCupMatches.csv')
pd.set_option('display.max_columns', None)
df
```

# *Task 1: Deletando os valores núlos*

```python
df = df.dropna()
df
```

# **Question 1:**
# **Quantas copas ja foram disputadas?**

```python
# Será necessário pegar as informações somente das "finais", pois nenhuma copa
# tem duas finais
finals = df.loc[df['Stage'] == 'Final', ['Year', 'City', 'Home Team Name',
                                'Home Team Goals', 'Away Team Goals',
                                'Away Team Name', 'Win conditions']]
finals

# Sabemos que o primeiro campeão mundial foi o Uruguai em 1930 com uma vitória de
# 4 x 2 em cima da Argentina

# Porém podemos observar que no ano de "2014" está escrito duas vezes e precisamos resolver isso
```

# *Task 2: Removendo os valores duplicados*

```python
# O ano de "2014" estava duplicado no nosso df, então será necessário remover os
# valores duplicados
finals = finals.drop_duplicates('Year')
finals
```

# *Task 3: Renomear os valores "Germany FR" para "Germany"*

```python
# Sabemos que Germany FR e Germany são a mesma seleção ("Alemanha")
# então não faz sentido manter dois nomes diferentes para a mesma seleção
# vamos renomear para que nosso gráfico fique perfeito

finals = finals.apply(lambda x: x.replace('Germany FR', 'Germany'))
finals
```

# *Task 4: Descobrindo quantos jogos terminaram em empates (Jogos que foram decididos nos penaltis ou na prorrogação)*

```python
# Os jogos que terminaram em empates entram com o placar igual, portanto na hora
# de plotar o gráfico não vai haver um campeão, então será necessário descobrir
# qual seleção venceu nos jogos que terminaram em empates no tempo normal

tie = finals.loc[finals['Home Team Goals'] == finals['Away Team Goals']]
tie
```

```python
# A var "home" vai pegar o primeiro resultado de cada row enquanto a var "away"
# vai pegar o segundo resultado
home = tie['Win conditions'].str[-7]
away = tie['Win conditions'].str[-3] 

for i in home:
  print(i)
```

```python
home_result = tie['Win conditions'].apply(lambda x: x.split()[-3][1])
# type(home_result)

# Convertendo o resultado para o tipo "int64" para que possa ser feito operações
# aritiméticas
home_result = pd.to_numeric(home_result[0:])
home_result
```

```python
away_result = tie['Win conditions'].apply(lambda x: x.split()[-1][0])

away_result = pd.to_numeric(away_result[0:])
away_result
```

# *Task 5: Atribuindo os valores de "win conditions" ao resultado original dos empates*

```python
tie['Home Team Goals'] = home_result
tie['Away Team Goals'] = away_result
tie
```

```python
# Excluindo os valores núlos que foram gerados
tie.dropna(inplace=True)
tie
```

# *Task 6: Concatenando o df "finals" ao df "tie" e criando um só df*

```python
# concatenando o df "finais" ao df "empates" e criando um só df
'''matches = pd.concat([finais, empates])
matches'''

# Removendo os resultados anteriores dos empates
'''matches = matches.drop_duplicates() # não deu certo pois apagou o resultado mais
recente e manteve o empate'''

matches = pd.concat([finals, tie])
matches
```

# *Task 7: Excluindo os jogos "antigos" onde o resultado dos empates é igual e substituindo pelo "novo" resultado*

```python
tie_matches = matches.loc[matches['Home Team Goals'] == matches['Away Team Goals']]
tie_matches
```

# **Question 2: Quantas copas cada país venceu ?**

```python
# Vamos descobrir quais dos jogos disputados o "home team" venceu
home_team = matches.loc[matches['Home Team Goals'] > matches['Away Team Goals'],
                             ['Year', 'Home Team Name']]

# Aproveitando.. vamos incluir também o resultado dos anos que não constam
# no dataframe
home_team = home_team.append({'Year': 2018.0, 'Home Team Name': 'France'}, ignore_index=True)
home_team = home_team.append({'Year': 2022.0, 'Home Team Name': 'Argentina'}, ignore_index=True)
```

```python
home_team.index = home_team['Year']
home_team
```

```python
# Vamos descobrir quantas vezes o "away team" venceu 
away_team = matches.loc[matches['Away Team Goals'] > matches['Home Team Goals'],
                       ['Year', 'Away Team Name']]

away_team.index = away_team['Year']
away_team
```

# *Task 8: Concatenando o resultado de "Home team" ao de "Away team"*

```python
# Precisamos fazer uma concatenação do df home_team e away_team, mas para que 
# a concatenação possa ser feita todas as colunas de ambos os dfs precisam ter
# o mesmo nome, então será necessário alterar o nome da coluna de um dos dfs, eu escolhi
# o df away_team

away_team.rename(columns={'Away Team Name': 'Home Team Name'}, inplace=True)
away_team

# o parâmetro inplace permite que as alterações feitas permanecam salvas 
```

```python
total_of_matches = pd.concat([home_team, away_team])
total_of_matches
```

# *Task 9: Ordenando os jogos por ordem cronológica*

```python
# Precisamos ordenar os índices pois depois da concatenação os anos não seguem 
# uma ordem cronológica.

total_of_matches = total_of_matches.sort_index()
total_of_matches
```

# *Task 10: Plotando um gráfico*

```python
import matplotlib.pyplot as plt

plt.plot(total_of_matches['Year'], total_of_matches['Home Team Name'], 'ro')
plt.title('WorldCup Matches', size=(30))
plt.ylabel('Team Names', size=(20))
plt.xlabel('Year', size=(20),)
plt.yticks(color='green')
plt.xticks(total_of_matches['Year'], rotation=90, color='blue')
plt.grid()
plt.show()
```
