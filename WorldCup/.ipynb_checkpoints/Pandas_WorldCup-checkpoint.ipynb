# **Question 1:**
# **Quantas copas ja foram disputadas?**

```python
# Coletando os valores da variável "Stage" onde sejam iguais à "Final"
finals = df.loc[df['Stage'] == 'Final', ['Year', 'City', 'Home Team Name',
                                'Home Team Goals', 'Away Team Goals',
                                'Away Team Name', 'Win conditions']]
finals
```


<p>Sabemos que o primeiro campeão mundial foi o Uruguai em 1930 com uma vitória de
4 x 2 em cima da Argentina. </p>

<p>Porém podemos observar que o ano de <b>2014</b> está duplicado e precisamos resolver isso.</p>

<h1><b>Removendo os valores duplicados.</b></h1>

```python
# Removendo o valor duplicado em 2014
finals = finals.drop_duplicates('Year')
finals
```

<p>Os valores que contém os nomes das seleções, existe dois valores diferentes para a mesma seleção:</p>
<ul>
    <li>Germany FR</li>
    <li>Germany</li>
</ul>
<p>Vamos renomeá-lo para que o pandas entenda que se trata da mesma seleção.</p>

```python
# Renomeando Germany FR para Germany
finals = finals.apply(lambda x: x.replace('Germany FR', 'Germany'))
finals
```

<h1><b> Quantos Jogos Terminaram em Empates ?</b></h1>
<p>No futebol quando se trata de jogos eliminatórios não é possível que um jogo termine em empate, nesse caso existe a <b>prorrogação</b> para que os times tenham a oportunidade de abrir liderança no placar, e mesmo assim se depois da prorrogação não houver um desempate então os times vão para o mata mata nos penaltis.</p>
<p>Portanto será necessário tratar esses valores onde em tempo normal os jogos terminaram em empates, para isso vamos analisar a variável "Win Conditions".</p>

```python
# Criando a variável de empate onde ambos os valores dos times foram iguais
tie = finals.loc[finals['Home Team Goals'] == finals['Away Team Goals']]
tie
```

<p>De todos os jogos até o momento presentes no nosso conjunto de dados, tivemos apenas dois empates em 1994 e 2006.</p>
<p>PS: Em 2022 foi decidido nos pentaltis entre França e Argentina, mas nesse conjunto de dados ainda não tem essa informação.</p>

```python
# A variável "home" vai receber o primeiro resultado da variável "Win Conditions" 
# Enquanto a variável "away" vai receber o segundo resultado
home = tie['Win conditions'].str[-7]
away = tie['Win conditions'].str[-3]
```

```python
home
```

```python
away
```

```python
# Salvando o resultado final para a seleção de casa
home_result = tie['Win conditions'].apply(lambda x: x.split()[-3][1])


# Convertendo o resultado para o tipo "float64" para que possa ser feito operações
home_result = home_result.astype("float64")
home_result
```

```python
# Salvando o resultado final para a seleção de fora
away_result = tie['Win conditions'].apply(lambda x: x.split()[-1][0])

# Convertendo para o tipo float
away_result = away_result.astype("float64")
away_result
```

<h1><b>Imputando os Valores onde o Resultado Foi Empate</b></h1>

```python
tie['Home Team Goals'] = home_result
tie['Away Team Goals'] = away_result
tie
```

```python
# Verificando se foi gerado algum valor nulo
tie.isnull().any()
```

```python
# Concatenando o DataFrame Principal ao DataFrame Empates
matches = pd.concat([finals, tie])
matches
```

```python
# Removendo do conjunto de dados as partidas que terminaram em empates
matches = matches.loc[matches['Home Team Goals'] != matches['Away Team Goals']]
matches
```

<p>Nos anos de 1994 e 2006 onde os jogos haviam terminados empatados agora possuem
o valor de "win condition" que foi o resultado final decidido na prorrogação ou 
no penaltis, agora temos de fato um campeão para os jogos.</p>

<h1><b>Quantas Copas Cada Seleção Venceu ?</b></h1>

```python
# Vamos descobrir quais dos jogos disputados o "home team" venceu
home_team = matches.loc[matches['Home Team Goals'] > matches['Away Team Goals'],
                             ['Year', 'Home Team Name']]

# Aproveitando.. vamos incluir também o vencedor (sem resultados) dos anos que não constam
# no dataframe
home_team = home_team.append({'Year': 2018.0, 'Home Team Name': 'France'}, ignore_index=True)
home_team = home_team.append({'Year': 2022.0, 'Home Team Name': 'Argentina'}, ignore_index=True)
```

```python
home_team.index = home_team['Year']
home_team
```

```python
# Vamos descobrir quantas vezes o "away team" venceu 
away_team = matches.loc[matches['Away Team Goals'] > matches['Home Team Goals'],
                       ['Year', 'Away Team Name']]

away_team.index = away_team['Year']
away_team
```

```python
# Vamos concatenar todos os resultados, mas para isso precisamos alterar o nome da variável
away_team.rename(columns={'Away Team Name': 'Home Team Name'}, inplace=True)
away_team
```

```python
total_of_matches = pd.concat([home_team, away_team])
total_of_matches
```

```python
# Ordenando os dados de acordo com o ano
total_of_matches = total_of_matches.sort_index()
total_of_matches
```

```python
# Melhorar a qualidade do gráfico
%config InlineBackend.figure_format = "svg"
```

```python
# Figura e eixo
fig, ax = plt.subplots(figsize=(6,3))

# Eixo x Year, eixo Y nome da seleção
ax.plot(total_of_matches["Year"], total_of_matches["Home Team Name"], "ro", color = "#000033");

# Rotacionando os dados do eixo x
plt.xticks(total_of_matches['Year'], rotation=90, color='#333333', size = 9);
# Alterando a cor do eixo y
plt.yticks(size = 9, color = "#004c00")

# Second War legend
fig.text(x = 0.27,
        y = 0.6,
        s = "Second\n   War",
        fontdict = {"family": "arial",
                    "size": 8,
                    "weight": "bold", 
                    "color": "#333333"},
                    alpha = 0.8)

# Título
fig.text(x = 0.14,
         y = 1,
         s = "World Cup Winners",
         fontdict = {"family": "arial",
                     "size": 16,
                     "weight": "bold",
                     "color": "#333333"},
        alpha = 0.8)

ax.grid()

fig.tight_layout()
plt.show()
```
