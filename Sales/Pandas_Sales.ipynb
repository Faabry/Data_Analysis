# **Importando as bibiotecas necessárias**

```python
import pandas as pd
```

```python
df = pd.read_csv('all_data.csv')
df
```

# *Task 1: Criar uma coluna "Month" com apenas os dois primeiros valores da coluna "Order Date"*

```python
# Peguei os primeiros 2 valores da coluna "Order Date"
df['Month'] = df['Order Date'].str[0:2] 
df
```

# *Task 2: Localizando e excluindo os valores núlos*

```python
# Criando uma variável que irá receber todas linhas e colunas com valores NaN
nan_df = df[df.isna().any(axis=1)]
nan_df

# 545 linhas serão excluidas
```

```python
# Excluindo os valores núlos
df = df.dropna(how='all')
df.head()
```

# *Task 3: Localizando e excluindo as linhas da coluna Month onde o valor seja "Or"*

```python
# Localizando as linhas da coluna Month onde o valor seja "Or"
order = df.loc[df['Month'] == 'Or']
order

# Essas colunas são inválidas para nossa operação, pois como na coluna "Order Date"
# não temos uma data e sim uma str consequentemente nossa coluna "Month" está com o valor
# "Or" ao invés de um número de mês
```

```python
# Excluindo do nosso df os valores "Or"
df = df[df['Month'] != 'Or']
df
```

# *Task 4: Converter os valores da coluna Month para o tipo integer*

```python
df['Month'] = df['Month'].astype('int32')
# A primeira vista não foi possível converter para int pois existem valores dentro do df
# que são núlos e precisam ser excluidos


# Na segunda tentativa de converter foi retornado o seguinte erro: 
# ValueError: invalid literal for int() with base 10: 'Or'

# Ou seja, foi necessário excluir as linhas da coluna Month onde o valor sera "Or"
```

# *Task 5: Converter os valores das colunas Quantity Order e Price Each de str para float*

```python
df['Price Each'] = pd.to_numeric(df['Price Each']) # Converter os valores das vendas para float
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered']) # Converter a quantidade de transações para int
```

# *Task 6: Criar uma coluna com o nome de cada cidade*

```python
# Basicamente estou criando uma coluna com o nome "City" que vai receber somente
# O indice[1] da coluna Purchase Address, uma vez que ele foi repartido em 3 partes
# Com a função split()
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1])
df.head()
```

# **Questão 1: Qual foi mês que teve mais vendas?**

```python
# Para descobrir qual foi o mês com mais vendas será necessário criar uma nova 
# Coluna que receba o valor final de cada transação efetuada

df['Sales'] = df['Quantity Ordered'] * df['Price Each']
# Recebemos o seguinte erro na primeira tentativa:
# TypeError: can't multiply sequence by non-int of type 'str'

# Foi necessário converter os valores das colunas para numeric

df
```

# *Task 7: Criar um groupby com a soma dos valores de cada mês*

```python
best_month = df.groupby('Month').sum()
best_month
```

```python
best_month.max() 

best_month.loc[best_month['Sales'] == 4613443.34]
# O melhor mês foi o de "Dezembro" com 28.114 Pedidos e $4.613.443.34 vendas 
```

# *Task 9: Plotando um gráfico com o faturamento mensal*

```python
import matplotlib.pyplot as plt

months = range(1, 13)

plt.bar(months, best_month['Sales'])
plt.xticks(months)
plt.xlabel('Months', size=(20))
plt.ylabel('Sales in $', size=(20))
plt.title('Result of a year Sales', size=(25))
plt.grid(True)
plt.show()
```

# **Questão 2: Qual foi a cidade com o melhor faturamento?**

# *Task 10: Criar um groupby com os nomes das cidades e seus respectivos valores somados*

```python
best_revenue = df.groupby('City').sum()
best_revenue
```

# *Task 11: Descobrir qual foi a cidade com o maior faturamento*

```python
best_revenue.max()
```

```python
best_revenue.loc[best_revenue['Sales'] == 8262203.91]

# A cidade com maior faturamento foi a de San Francisco com 50.239 pedidos 
# E $8.262.203.91 de faturamento
```

# *Task 12: Plotar um gráfico com o faturamento anual de cada cidade*

```python
# Antes de criar um plot precisamos criar o parametro que será exibido no gráfico

# Criando uma variável que irá receber o valor de cada cidade individualmente
cities = [city for city, i in df.groupby('City')]
cities

plt.bar(cities, best_revenue['Sales'])
plt.xticks(cities, rotation='vertical', size=10)
plt.title('Yearly Sales of Each City', size=(25))
plt.xlabel('City Name', size=(20))
plt.ylabel('Sales in $', size=(20))
plt.grid(True)
plt.show()
```

# *Task 13: Criando uma coluna "Hour" que irá receber a hora de cada transação*

```python
# Primeiramente é necessário converter a coluna 'Order Date' para o tipo datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
# Criando a coluna 'Hour'
df['Hour'] = df['Order Date'].dt.hour
df
```

# *Task 14: Agrupando todas as horas*

```python
hour = [hour for hour, df in df.groupby('Hour')]
hour
```

# *Task 15: Plotando um gráfico*

```python
# Ao criar o plot é necessário criar um novo groupby de df['Hour']
plt.plot(hour,  df.groupby(['Hour']).count(), linestyle=':', color='green')
plt.title('Flow Sales During a Day', size=(20), color='blue')
plt.ylabel('Number of Orders', size=(15))
plt.xlabel('Hour', size=(15))
plt.xticks(hour, color='red')
plt.grid()
plt.show()

# O melhor fluxo é entre 11hrs e 19hrs
```
