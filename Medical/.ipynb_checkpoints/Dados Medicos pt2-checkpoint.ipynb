<h2>1- Total de Medicamentos Por Faixa Etária</h2>

```python
# Cores para os gráfico
colors = ["#404040", "#404040", "#404040", "#00004c", "#404040"]
```

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(20, 8))

# Gráfico de barras
ax = sns.barplot(x = df.age, y = df.num_med, estimator = (np.sum), data = df, palette= colors)

# Removendo as labels
plt.yticks([])
plt.ylabel("")
plt.xticks(size = 15)
plt.xlabel("")

# Título
fig.text(x = 0.13,
         y = 0.9,
         s = "Total de Medicamentos Por Faixa Etária",
         fontdict = {"size": 20,
                     "weight": "bold",
                     "color": "#333333"},
         alpha = 0.6)

# Insight
fig.text(x = 0.74,
         y = 0.79,
         s = "Pacientes entre $\\bf{70-80}$ anos\nConsomem mais medicamentos",
         fontdict = {"size": 15,
                     "color": "#00004c"},
         alpha = 0.6)

# Rótulo dos dados
for i in ax.patches:
    ax.annotate(f"{i.get_height():.0f}", (i.get_x() + 0.25, i.get_height()),
                color = "black", ha = "center", size = 15)

# Removendo as bordas
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# Exibir
plt.show()
```

<h2>2- Existe Alguma Diferença Entre Readimissões Por Gênero?</h2>

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(20,8))

# Gráfico countplot
ax = sns.countplot(x = [("Sexo Masculino" if x == 1 else "Sexo Feminino") for x in df["gender"]],
                   hue = "readmitted",
                   data = df)

# Legenda
ax.legend(title = "Paciente Readimitido", labels = ["Não", "Sim"],
          loc='best', bbox_to_anchor=(0.45, 0.45, 0.5, 0.5),
          fontsize="large", edgecolor="black", title_fontsize=13)

# Removendo os labels
plt.ylabel("")
plt.yticks([])
plt.xticks(size = 15)

# Rótulo de dados
for c in ax.patches:
    ax.annotate(f"{c.get_height():.0f}", (c.get_x() + 0.2, c.get_height()),
                color = "black", ha = "center", size = 15)

# Título
fig.text(x= 0.16,
         y= 0.9,
         s= "Total De Readimissões Por Gênero",
         fontdict={"size": 20, "weight": "bold"},
         alpha= 0.6)

# Insight
fig.text(x= 0.33,
         y= 0.72,
         s= "Pacientes do sexo $\\bf{feminino}$\nTem mais readimissões",
         fontdict={"size": 15,
                   "color": "#00004c"},
         alpha= 0.6)

# Removendo as bordas
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# Exibir
plt.show()
```

<h2>3- Total de Readmissões Por Faixa Etária</h2>

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(20, 8))

# Gráfico de barras
# Barplot contabiliza somente os valores "True (1)", portanto não é necessário informar os dados 
# de pacientes que NÃO foram readmitidos False(0)
ax = sns.barplot(x=df.age, y=df.readmitted, estimator=np.sum, palette=colors)

# Removendo os labels
plt.xticks(size = 15)
plt.xlabel("")
plt.yticks([])
plt.ylabel("")

# Título
fig.text(0.14,
         0.9,
         "Readmissões Por Faixa Etária",
         fontdict={"size": 20,
                   "weight": "bold"},
         alpha=0.6)

# Insight
fig.text(0.75,
         0.77,
         "Pacientes entre $\\bf{70-80}$ anos\nTiveram mais readmissões.",
         fontdict={"size": 18,
                   "color": "#00004c"},
         alpha=0.6)

# Rótulo de dados
for i in ax.patches:
    ax.annotate(f"{i.get_height():.0f}", (i.get_x() + 0.25, i.get_height()),
                color="black", ha="center", size=18)
    
# Removendo as bordas   
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)

# Exibir
plt.show()
```

<h2>4- Percentual de Readmissões & Não Readmissões de Pacientes Por Faixa Etária.</h2>

```python
# Calcula o percentual por faixa etária
percentual = pd.crosstab(index=df.age, columns=df.readmitted, margins=True, normalize="index") * 100
```

```python
percentual
```

```python
# Gráfico a partir do dataframe
fig = percentual.plot(kind="bar",
                figsize=(20, 8),
                width=0.8,
                edgecolor="black")

# Removendo os labels
plt.xticks(rotation=0, size=15)
plt.xlabel("")
plt.ylabel("")
plt.yticks([])

# Rótulo de dados
for i in fig.patches:
    fig.annotate(f"{i.get_height():.0f}%", (i.get_x() + 0.2, i.get_height()),
                 color="black", ha="center", size=15, va="bottom")

# Removendo as bordas
fig.spines["right"].set_visible(False)
fig.spines["top"].set_visible(False)
fig.spines["bottom"].set_visible(False)
fig.spines["left"].set_visible(False)

# Insight
fig.text(x= -1.66,
         y=58,
         s="Pacientes entre $\\bf{0-50}$ anos\nTem maior percentual\nde $\\bf{Não}$ readmissão",
         fontdict={"size": 15,
                   "color": "#094a85"},
         alpha=0.6)

# Insight
fig.text(x= -1.66,
         y=48,
         s="Pacientes entre $\\bf{70-80}$ anos\nTem maior percentual\nde $\\bf{readmissão}$",
         fontdict={"size": 15,
                   "color": "#be4c00"},
         alpha=0.6)

# Legenda
plt.legend(title="Paciente Readmitido", labels=["Não", "Sim"])

# Exibir
plt.show()
```

<h2>5- Total de Readmissões por Gênero, Idade, Tipo de Admissão.</h2>

```python
# Paleta de cores
palette = ["#5486b4", "#d4a2bc"]
```

```python
# Background
sns.set(style="white")

# Catplot kind bar
ax = sns.catplot(x="age",
                 y="readmitted",
                 hue="gender",
                 col="admission_type_id",
                 estimator=np.sum,
                 data=df,
                 kind="bar",
                 palette=palette,
                 ci=None,
                 legend=False)

# Legenda
plt.legend(title="Gênero", labels=["Feminino", "Masculino"])

# Label eixo Y, titulo com nomes dos registros da variável "admission_type_id"
(ax.set_axis_labels("", "Total de Readmissões").set_titles("{col_name}"))

# Exibir
plt.show()
```

<p>Mulheres entre 70-80 anos tiveram mais readmissões por emergência.</p>

<h2>6- Total de Readmissões/Não Readmissões por Gênero e Raça.</h2>

```python
fig, ax = plt.subplots(figsize=(20,12))

ax = sns.countplot(y=df.race, hue=df.readmitted, data=df)

plt.yticks(size=15)
plt.ylabel("")
plt.xlabel("")

fig.text(x=0.123,
         y=0.9,
         s="Readmissões por Raça",
         fontsize=25)

ax.legend(title="Readmitido", labels=["Não", "Sim"], title_fontsize=15, fontsize=15, loc="right")

plt.show()
```

```python
# Valores únicos da variável race
df["race"].unique()
```

```python
# dataframe temporário
df_temp = df
```

```python
# Recategorizando a variável para o tipo str
df_temp["readmitted"] = df_temp["readmitted"].map({0: "Não Readmitido", 1: "Readmitido"})
```

```python
df_temp["readmitted"].head()
```

```python
def countplot(x, hue, **kwargs):
    sns.countplot(x = x, hue = hue, **kwargs)
```

```python
sns.set(style="white", context="talk")

grid = sns.FacetGrid(data=df_temp, col="readmitted", height=9, aspect=1)

fig = grid.map(countplot, "race", "gender", palette="deep")

(fig.set_axis_labels("", "Total")
 .set_xticklabels(["Caucasian", "AfricanAmerican", "Other", "Asian", "Hispanic"])
 .set_titles("{col_name}"))

plt.legend(title="Gênero", labels=["Feminino", "Masculino"])
plt.show()
```

<h2>8- Proporção de Readmissões pelo Nº de Visitas Antes da Alta.</h2>

```python
percent = pd.crosstab(df["number_inpatient"], df["readmitted"], normalize="index") * 100
percent
```

```python
fig = percent.plot(kind="bar", width=0.8, figsize=(20,10), edgecolor="black", color=["b", "r"])

for i in fig.patches:
    fig.annotate(f"{i.get_height():.0f}%", (i.get_x() + 0.2, i.get_height()),
                 color="black", ha="center", size=14, va="bottom")

fig.legend(title="Readmitido", labels=["Não", "Sim"])
    
fig.text(x= - 0.6,
         y=110,
         s="Proporção de Readmissões por Nº de Visitas Antes da Alta.",
         fontsize=20)

plt.yticks([])
plt.xlabel("")
plt.show()
```

<p>A medida que o paciente recebe mais visitas do médico ele tende a ser readmitido novamente.</p>
