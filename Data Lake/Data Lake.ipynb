<h1>Pré-Processamento e Análise</h1>

<h3>Número de Posts por Categoria</h3>

```python
# Agregação para calcular o nº de posts em cada categoria
category = posts.aggregate([{"$group": {"_id": "$Category", "count": {"$sum": 1}}},
                           {"$sort": {"_id.Category": 1, "count": -1}}])

# $group = agrupamento, nesse caso pelo _id em relação à Categoria
# count = contar pela soma de cada item ($sum: 1)
# $sort = ordenar pelo _id.Category para cada item
```

```python
print(category)
```

```python
# Convertendo o resultado da agregação em um dataframe
df = pd.DataFrame(list(category))
```

```python
df
```

```python
# Definindo os eixos X e Y
x = df["_id"]
y = df["count"]
```

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(10, 5))

# Gráfico de barras
ax = plt.bar(x=x, height=y, data=df)

ax = plt.gca()

# Rótulo de dados
for i in ax.patches:
    ax.annotate(f"{i.get_height()}", (i.get_x() + 0.3, i.get_height()),
                color = "black", ha="center", va="bottom")

# Removendo os labels do eixo Y
plt.yticks([])

# Título
fig.text(x=0.125,
         y=0.9,
         s="Número de Posts Por Categoria",
         fontsize=15)

# Exibir
plt.show()
```

<h3>Número de Posts Públicos e Privados da Categoria Politics</h3>

```python
# Criando a agregação dos dados
politics = posts.aggregate([{"$match": {"Category": "politics"}},
                           {"$group": {"_id": "$status", "count": {"$sum": 1}}}])

# $match = criar o casamento dos dados por categoria onde seja "politics"
# $group = fazer um agrupamento pelo _id em relaçao à "status" (public/private)
# count =  contar os registros somando um por um ($sum)
```

```python
# Convertendo o resultado da agregação para um DataFrame
politics_df = pd.DataFrame(list(politics))
```

```python
politics_df
```

```python
plt.pie(x=politics_df["count"].values, data=politics_df, labels=["Public", "Private"], startangle=90,
        autopct='%1.0f%%', colors=["#cedce9", "#3b75a9"]);
```

<h3>Qual Ano teve Mais Posts? </h3>

```python
# Update_many (cria novas variáveis, neste caso duas)
datalake.posts.update_many({}, {"$set": {"hour": 0}});
datalake.posts.update_many({}, {"$set": {"year": 0}});
```

```python
posts.find_one()
```

```python
# Para cada documento nas coleções, vai atualizar os atributos hour e year usando a função split
for document in posts.find({}):
    posts.update_one({"_id": document["_id"]},
                    {"$set": {"hour": document["creationDate"].split("T")[1].split(":")[0],
                              "year": document["creationDate"].split("T")[0].split("-")[0]}})
```

```python
# Retorna 1 item
posts.find_one()
```

```python
# Agregação
posts_yr = posts.aggregate([{"$group": {"_id": "$year", "count": {"$sum": 1}}},
                            {"$sort": {"_id.year": 1, "count": 1}}])

# Agrupando e fazendo a agregação pelo year, ordenando
```

```python
# Colocando o resultado da agregação em um DataFrame
df_year = pd.DataFrame(list(posts_yr))
```

```python
df_year
```

```python
# Definindo os eixos X Y
x = df_year["_id"]
y = df_year["count"]
```

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(10, 5))

# Gráfico de linha com pontos de scatter
ax = sns.lineplot(data=df_year, x=x, y=y)
ax = plt.scatter(x=x, y=y)

# Grid 
plt.grid()

# Removendo os labels
plt.xlabel("")
plt.ylabel("")

# Título
fig.text(x=0.125,
         y=0.9,
         s="Qtd Posts Por Ano",
         fontsize=20)

# Exibir
plt.show()
```

<h3>Comprimento dos Títulos dos Posts</h3>

```python
# Criando uma nova variável com o comprimento dos títulos (inicializando em Zero)
datalake.posts.update_many({}, {"$set": {"lenght_title": 0}});
```

```python
posts.find_one()
```

```python
for title in posts.find({}):
    posts.update_one({"_id": title["_id"]},
                     {"$set": {"lenght_title": len(title["title"].split(" "))}})
```

```python
posts.find_one()
```

```python
# Agrupamento e ordenação
len_title = posts.aggregate([{"$group": {"_id": "$lenght_title", "count": {"$sum": 1}}},
                             {"$sort": {"_id.lenght_title": 1, "count": -1}}])
```

```python
len_title
```

```python
df_titles = pd.DataFrame(list(len_title))
```

```python
df_titles
```

<h3>Qual Categoria tem Descrição Mais longa (acima de 60 caracteres)? </h3>

```python
# Criando a variável com a informação de tamanho da descrição
datalake.posts.update_many({}, {"$set": {"length_desc": 0}})
```

```python
# Agregando as informações do comprimento da descrição
for post in posts.find({}):
    posts.update_one({"_id": post["_id"]},
                     {"$set": {"length_desc": len(post["description"].split(" "))}})
```

```python
# Filtrando, agregando e ordenando
catg_desc = posts.aggregate([{"$match": {"length_desc": {"$gte": 60}}},
                             {"$group": {"_id": "$Category", "count": {"$sum": 1}}},
                             {"$sort": {"_id.tags": 1, "count": -1}}])
```

```python
# Criando o DataFrame
df_catg_desc = pd.DataFrame(list(catg_desc))
```

<h3>A Qtd de Post Varia de Acordo com A Hora do Dia? </h3>

```python
# Agrupando por categoria e hora e ordenando por hora
post_hr = posts.aggregate([{"$group": {"_id": ["$hour", "$Category"], "count": {"$sum": 1}}},
                           {"$sort": {"_id.hour": 1, "count": -1}}])
```

```python
final_df = pd.DataFrame(post_hr)
```

```python
final_df
```

```python
# Separando as informações de hora e categoria
split_id = pd.DataFrame(final_df["_id"].to_list(), columns=["hour", "category"])
```

```python
split_id
```

```python
df = pd.concat([split_id, final_df["count"]], axis=1)
```

```python
df
```

```python
result = df.groupby(["hour", "category"])["count"].sum()
```

```python
result
```

```python
# resultado alternativo
for key, value in df.groupby(["category"]):
    print(value)
```

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(10, 5))

# percorrendo os elementos dentro do groupby
for key, value in df.groupby(["category"]):
    ax = value.plot(ax = ax, kind = "line", x = "hour", y = "count", label = key)

# Legenda
plt.legend(loc = "best")

# Exibir
plt.show()
```
