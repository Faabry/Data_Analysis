<h1> Exploratory Analysis </h1>

```python
columns_to_convert = ['Density\n(P/Km2)', 'Agricultural Land( %)', 'Land Area(Km2)',
                      'Birth Rate', 'Co2-Emissions', 'Forested Area (%)',
                      'CPI', 'CPI Change (%)', 'Fertility Rate', 'Gasoline Price', 'GDP',
                      'Gross primary education enrollment (%)', 'Armed Forces size',
                      'Gross tertiary education enrollment (%)', 'Infant mortality',
                      'Life expectancy', 'Maternal mortality ratio', 'Minimum wage', 
                      'Out of pocket health expenditure', 'Physicians per thousand', 
                      'Population', 'Population: Labor force participation (%)', 
                      'Tax revenue (%)', 'Total tax rate', 'Unemployment rate', 'Urban_population']
```

```python
df[columns_to_convert] = df[columns_to_convert].applymap(lambda x: float(str(x) \
                                               .replace(',', '') \
                                               .replace('$', '') \
                                               .replace('%', '')))
```

```python
# Missing values?
for col in df.columns:
    if df[col].isna().sum() > 0:
        print(f"{col}: {df[col].isna().sum()} missing values.")
        print(f"Which represents {round(df[col].isna().sum() / len(df), 2)}% of the whole dataset.")
        print("-=" * 30)
```

<h1> Descriptive Analysis </h1>

```python
# Creating a numerical variable to receive the float columns
numerical_columns = df.select_dtypes(include = ["float64"])
numerical_columns.describe()
```

```python
# Printing the mean and median because the representation is in cientific notation
print(f"Land Area(Km2) mean  : {round(df['Land Area(Km2)'].mean(), 2)}")
print(f"Land Area(Km2) median: {df['Land Area(Km2)'].median()}")
```

<p> The difference between the mean and the median on Land Area(Km2) is huge, let's check it out if there's any invalid outlier.</p>

```python
# Increasing the graphic quality
%config InlineBackend.figure_format = "svg"
```

```python
ax, fig = plt.subplots(figsize=(6, 30))

# Create a boxplot
ax = sns.scatterplot(y=df["Country"], x=df["Land Area(Km2)"], data=df)

# Set labels and title
plt.xlabel("Land Area (Km²)")
plt.ylabel("")
ax.set_title("Distribution of Land Area by Country")

plt.grid(visible=True)

# Rotate x-axis labels for better readability
plt.xticks(size=15)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Imputation with median because of the outlier values
# Looking for the NaN values
land_area = df[df["Land Area(Km2)"].isna()].index

land_median = df["Land Area(Km2)"].median()
df["Land Area(Km2)"][land_area] = land_median 
```

<h1> Q1: The top 10 Countries with Biggest Land Area </h1>

```python
top_10 = df.groupby(["Country"])["Land Area(Km2)"].agg(["sum"]).sort_values(by="sum",
                                                                   ascending=False).head(10)
```

```python
top_10.reset_index
top_10
```

```python
colors = ["#00004c"]

for i in range(0, 3):
    colors.append("#1919ff")
    
for j in range(0, 2):
    colors.append("#4c4cff")
    
for y in range(0, 4):
    colors.append("#7f7fff")
```

```python
fig, ax = plt.subplots(figsize=(7.5, 3))

ax = sns.barplot(x=top_10.index, y=top_10["sum"], palette=colors)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points',
                size = 7)
    

fig.text(y = 0.95,
         x = 0.13,
         s = "Top 10 Countries With the Biggest Land Area(Km2)",
         fontdict = {"family": "bookman old style",
                     "size": 15})
    
plt.xticks(size = 6)
plt.xlabel("")
plt.yticks([])
plt.ylabel("")


ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

plt.show()
```

<p> The top 10 countries with the highest and lowest GDP </p>
<p> As we know the GDP variable has 2 missing values, so let's check it out to find out the best way to make an imputation</p>

```python
df["GDP"].describe()
```

```python
print(f"GDP mean  : {round(df['GDP'].mean(), 1)}")
print(f"GDP median: {df['GDP'].median()}")
```

<p>There's a huge difference between the mean and the median too, it occurs because the countries has it's resources and the economy of each country is totally different.</p>

```python
# Imputation on the missing values the median
apply = df[df["GDP"].isna()].index
df["GDP"][apply] = df["GDP"].median()
```

```python
# Top 10 highest country GDP
highest = df.groupby("Country")["GDP"].agg(["sum"]).sort_values(by="sum",
                                                      ascending = False).head(10)
```

```python
# Top 10 lowest country GDP
lowest = df.groupby("Country")["GDP"].agg(["sum"]).sort_values(by="sum",
                                                      ascending = True).head(10)
```

```python
# ax, fig = plt.subplots(figsize=(6, 4))

plt.figure(figsize=(6, 4))

plt.subplot(211)
ax1 = sns.barplot(y=highest.index,
            x=highest["sum"],
            orient="h")
plt.title("Top 10 Highest Countries GDP")
plt.yticks(size = 7)
plt.ylabel("")
plt.xlabel("")
plt.xticks([])

ax1.spines["right"].set_visible(False)
ax1.spines["left"].set_visible(False)
ax1.spines["top"].set_visible(False)
ax1.spines["bottom"].set_visible(False)

for i, v in enumerate(highest['sum']):
    ax1.text(v + 10, i, str(v),
             color='black',
             ha='left',
             va='center',
             size= 4 )


plt.subplot(212)
ax2 = sns.barplot(y=lowest.index,
            x=lowest["sum"],
            orient="h")
plt.title("Top 10 Lowest Countries GDP")
plt.yticks(size = 7)
plt.ylabel("")
plt.xlabel("")
plt.xticks([])

ax2.spines["right"].set_visible(False)
ax2.spines["left"].set_visible(False)
ax2.spines["top"].set_visible(False)
ax2.spines["bottom"].set_visible(False)


for i, v in enumerate(lowest['sum']):
    ax2.text(v + 10, i, str(v),
             color='black',
             ha='left',
             va='center',
             size= 4 )


plt.tight_layout()
plt.show()
```
